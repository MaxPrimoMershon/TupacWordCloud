```python
import pandas as pd
import urllib.request
import numpy as np
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
from nltk.tokenize import TweetTokenizer
import string
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
from os import path
import os
```

```python
# Returns consumable html of the webpage for the album passed

def gettracklisthtml(url):
    page = requests.get(url)
    html = BeautifulSoup(page.text, "html.parser")
    return html
```

```python
# Returns a list of all the tracks on an album

def gettracklist(html):
    tracklist = html.findAll("h3", {"class": "chart_row-content-title"})
    x = 0 
    for i in tracklist:
        tracklist[x] = tracklist[x].getText().strip("\n")
        tracklist[x] = tracklist[x].strip("\n              Lyrics")
        x +=1
    tracklist = pd.DataFrame(tracklist, columns=["Tracks"])
    tracklistlinks = html.findAll("h3", {"class": "chart_row-content-title"})
    x = 0 
    for i in tracklistlinks:
        tracklistlinks[x] = tracklistlinks[x].parent
        tracklistlinks[x] = tracklistlinks[x]["href"]
        x +=1
    tracklist["Link"] = tracklistlinks
    albumnumber = html.findAll("span", {"class": "chart_row-number_container-number"})
    x = 0 
    for i in albumnumber:
        albumnumber[x] = albumnumber[x].getText().strip("\n")
        x +=1
    tracklist["Track Number"] = albumnumber
    tracklist = tracklist.loc[tracklist["Track Number"] != ""]
    tracklist.index = tracklist["Track Number"]
    tracklist.drop(["Track Number"], axis=1, inplace=True)

    return tracklist
```

```python
#returns a list of tracks and lyrics on each track

def lyricscount(tracks):
    tracks["Lyrics"] = ""
    x = 0
    for i in tracks.index:
        url = tracks.iloc[x]["Link"]
        page = requests.get(url)
        html = BeautifulSoup(page.text, "html.parser")
        lyrics = html.find("div", {"class":"lyrics"}).getText().replace("\n", " ")
        tracks.iat[x,2] = lyrics
        x += 1
    return tracks
```

```python
# Aggregates the texts of the lyrics in all tracks and tokenizes the words

def lyricstext(albumdata):
    lyrics_all = ""
    for rownum, row in albumdata.iterrows():
        lyrics_all += albumdata.loc[rownum, "Lyrics"]

    lyrics_all = lyrics_all.lower()
    lyrics_all = lyrics_all.replace(" ("," ")
    lyrics_all = lyrics_all.replace(") "," ")
    lyrics_all = lyrics_all.replace("? "," ")
    lyrics_all = lyrics_all.replace("! ","")
    lyrics_all = lyrics_all.replace("nigga","n****")
    tokens = lyrics_all.split()
    stopwords = nltk.corpus.stopwords.words('english')
    punct = string.punctuation
    tokens = [w for w in tokens if w not in punct]
    tokens = [w for w in tokens if w not in stopwords]
    tokens = [w for w in tokens if "[" not in w]
    tokens = [w for w in tokens if "]" not in w]

    return tokens
```

```python
url = "https://genius.com/albums/2pac/All-eyez-on-me"
html = gettracklisthtml(url)
tracks = gettracklist(html)
album_tracks = lyricscount(tracks)
album_tracks
```

```python
lyrics = lyricstext(album_tracks)
lyrics = ' '.join(lyrics)
lyrics
```

```python
def grey_color_func(word, font_size, position,orientation,random_state=None, **kwargs):
    return "hsl(0, 0%%, %d%%)" % np.random.randint(0, 1)

d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()
mask = np.array(Image.open(path.join(d, "tupac3.jpg")))

wc = WordCloud(max_words=100000, mask=mask, random_state=5, background_color="white").generate(lyrics)
image_colors = ImageColorGenerator(mask)
default_colors = wc.to_array()

plt.figure(figsize=[15,15])
plt.imshow(wc.recolor(color_func=grey_color_func), interpolation="bilinear")
plt.axis("off")
plt.show()
wc.to_file("illmatic_wordcloud.png")
```
